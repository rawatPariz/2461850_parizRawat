```python
#2 Composite Development Score
hdi_south_asia = hdi_south_asia.copy()
le_min = hdi_south_asia['life_expectancy'].min()
le_max = hdi_south_asia['life_expectancy'].max()
hdi_south_asia.loc[:, 'life_exp_index'] = (hdi_south_asia['life_expectancy'] - le_min)/ (le_max - le_min)

gni_min = hdi_south_asia['gross_inc_percap'].min()
gni_max = hdi_south_asia['gross_inc_percap'].max()
hdi_south_asia.loc[:, 'gni_index'] = (hdi_south_asia['gross_inc_percap'] - gni_min)/ (gni_max - gni_min)

hdi_south_asia.loc[:, 'composite_score'] = (0.30 * hdi_south_asia['life_exp_index'] + 0.30 * hdi_south_asia['gni_index'])
south_asia_2022 = hdi_south_asia[hdi_south_asia['year'] == 2022]
composite_ranking = south_asia_2022.sort_values(by='composite_score', ascending=False)
composite_ranking[['country', 'composite_score']]
```

```python
#3 Outlier Detection

def detect_outliers(series):
  Q1 = series.quantile(0.25)
  Q3 = series.quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  return(series < lower_bound) | (series > upper_bound)
```

```python
south_asia_da_2022 = hdi_south_asia[hdi_south_asia['year'] == 2022].copy()

south_asia_da_2022['hdi_outlier'] = detect_outliers(south_asia_da_2022['hdi'])
south_asia_da_2022['gni_oulier']  = detect_outliers(south_asia_da_2022['gross_inc_percap'])

south_asia_da_2022['outlier'] = south_asia_da_2022['hdi_outlier'] | south_asia_da_2022['gni_oulier']
south_asia_da_2022[['country', 'hdi', 'gross_inc_percap', 'outlier']]
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    south_asia_da_2022[~south_asia_da_2022['outlier']]['gross_inc_percap'],
    south_asia_da_2022[~south_asia_da_2022['outlier']]['hdi'],
    label='Normal',
    alpha=0.7
)

plt.scatter(
    south_asia_da_2022[south_asia_da_2022['outlier']]['gross_inc_percap'],
    south_asia_da_2022[south_asia_da_2022['outlier']]['hdi'],
    label='Outliers',
    marker='o'
)

plt.xlabel('GNI per Capita')
plt.ylabel('Human Development Index (HDI)')
plt.title('Outlier Detection: HDI vs GNI per Capita (South Asia, 2022)')
plt.legend()
plt.grid(True, linestyle='-', alpha=0.9)
plt.show()

"Using the 1.5 × IQR rule to detect outliers among South Asian countries in 2022 found no outliers. This is expected because the region has few countries and their HDI and GNI per capita values are fairly similar."
```
